# Data appendix

Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.backends.backend_pdf import PdfPages
```

Get the data

```python
data = pd.read_csv("../Analysis_Data/vacuum_cleaner_statistics_podlaskie.csv")
```

Missing values

```python
n_missing = {col: list(data[col].isnull().values).count(True) for col in data.columns}
n_all = len(data)
txt = "\n".join([f"Name: {col}\nMissing values: {n_missing[col]}:{n_all}\n" for col in data.columns])
```

Basic summary statistics

```python
stat = {}
variables = ["Dni od zakupu", "Wiek", "Ocena"]

for var in variables:
    vals = [v for v in data[var].values if not pd.isna(v)]
    var_stat = {}
    var_stat["Mean"] = round(np.mean(vals), 4)
    var_stat["Std"] = round(np.std(vals), 4)
    var_stat["Min"] = np.min(vals)
    var_stat["25th percentile"] = np.percentile(vals, 25)
    var_stat["Median"] = np.percentile(vals, 50)
    var_stat["75th percentile"] = np.percentile(vals, 75)
    var_stat["Max"] = np.max(vals)
    stat[var] = pd.DataFrame(data=var_stat, index=["Statistics"])

fig_stat, ax_stat = plt.subplots(3, 1, figsize=(20, 6))
fig_stat.suptitle("Basic summary statistics for vacuum_cleaner_statistics_podlaskie.csv data", fontsize=20, y=1.1)

for i, var in enumerate(variables):
    ax_stat[i].axis('off')
    ax_stat[i].axis('tight')
    ax_stat[i].set_title(var, fontsize=18)
    var_stat_table = ax_stat[i].table(cellText=stat[var].values, colLabels=stat[var].columns, loc='center', cellLoc='center')
    var_stat_table.set_fontsize(18)
    var_stat_table.scale(1, 2)

plt.show()
```

Frequency tables

```python
freq = {}
variables = ["Marka", "Pe"]

for var in variables:
    freq[var] = pd.DataFrame(data={k: [v] for k, v in data[var].value_counts().to_dict().items()}, index=["Count"])

fig_table, ax_table = plt.subplots(2, 1, figsize=(20, 8))
fig_table.suptitle("Frequency tables for vacuum_cleaner_statistics_podlaskie.csv data", fontsize=20, y=1.1)

for i, var in enumerate(variables):
    ax_table[i].axis('off')
    ax_table[i].axis('tight')
    ax_table[i].set_title(var, fontsize=18)
    freq_table = ax_table[i].table(cellText=freq[var].values, colLabels=freq[var].columns, loc='center', cellLoc='center')
    freq_table.set_fontsize(18)
    freq_table.scale(1, 2)
    
plt.show()
```

Bar charts

```python
fig_chart, ax_chart = plt.subplots(3, 2, figsize=(20, 30))
fig_chart.tight_layout(pad=12.0)
fig_chart.suptitle("Variables' frequency for vacuum_cleaner_statistics_podlaskie.csv data", fontsize=20, y=1.0)
ax_chart[-1, -1].axis('off')

for i, col in enumerate(data):
    bins_cnt = data[col].value_counts().to_dict()
    
    if col == "Dni od zakupu":
        bins = ["<5", "5-9", "10-14", ">14"]
        cnt = [0] * 4
        
        for b in bins_cnt:
            if b < 5:
                cnt[0] += bins_cnt[b]
            elif 5 <= b < 10:
                cnt[1] += bins_cnt[b]
            elif 10 <= b < 15:
                cnt[2] += bins_cnt[b]
            else:
                cnt[3] += bins_cnt[b]
                
    elif col == "Wiek":
        bins = ["<20", "20-39", "40-59", ">60"]
        cnt = [0] * 4
        
        for b in bins_cnt:
            if b < 20:
                cnt[0] += bins_cnt[b]
            elif 20 <= b < 39:
                cnt[1] += bins_cnt[b]
            elif 40 <= b < 59:
                cnt[2] += bins_cnt[b]
            else:
                cnt[3] += bins_cnt[b]
        
    else:
        bins = [b for b in bins_cnt]
        cnt = [c for c in bins_cnt.values()]
        
    x, y = i//2, i%2
    
    ax_chart[x, y].bar(bins, cnt, tick_label=bins)
    ax_chart[x, y].set_title(col)
    ax_chart[x, y].set_xlabel("Variable")
    ax_chart[x, y].set_ylabel("Count")
    ax_chart[x, y].set_xticklabels(bins, rotation=30)
    
plt.show()
```

Save the missing values summary to a .txt file.

```python
DST = "../Analysis_Data/Data_Appendix_Output/"

with open(f"{DST}data_appendix_report.txt", "w") as f:
    f.write(txt)
```

Save the basic summary statistics tables, frequency tables and bar charts to .pdf files.

```python
with PdfPages(f"{DST}data_appendix_statistics.pdf") as pdf:
    pdf.savefig(fig_stat)
    pdf.savefig(fig_table)
    pdf.savefig(fig_chart)
```
