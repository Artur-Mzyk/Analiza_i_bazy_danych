***Data appendix***

Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Data description

```python
DIR = "../Analysis Data/"
survey = pd.read_csv(f"{DIR}earthquake_data_modified_column_names.csv")
print(survey.describe(include='all'))
```

Frequency tables and bar charts

```python
_, ax = plt.subplots(4, 3, figsize=(20, 15))

for idx, name in enumerate(survey.columns):
    freq_table = survey.value_counts(subset=[name], sort=False).sort_index()
    print(f"{freq_table}\n")      
    freq_table.plot(kind='bar', ax=ax[idx//3, idx%3])
    
ax[-1, -1].axis('off')
plt.show()
```
