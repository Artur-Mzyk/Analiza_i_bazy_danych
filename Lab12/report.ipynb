# Tworzenie bazy danych w SQLAlchemy

Potrzebne pakiety

```python
from sqlalchemy import create_engine, Column, Integer, SmallInteger, String, Date, Float, Text, ForeignKey, Table
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy_schemadisplay import create_schema_graph

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Połączenie z bazą danych

```python
try:
    db_string = "postgresql://postgres:postgres@localhost:5432"

    with create_engine(db_string, isolation_level="AUTOCOMMIT").connect() as connection:
        connection.execute("CREATE DATABASE lab12")
        
except:
    db_string = "postgresql://postgres:postgres@localhost:5432/lab12"
    engine = create_engine(db_string)
```

Uruchomienie automatycznego procesu mapowania istniejących tabel w bazie danych do klas.

```python
Base = declarative_base()
```

Utworzenie struktur tabel

```python
class User(Base):
    __tablename__ = "users"
    id = Column(Integer, primary_key=True)
    email = Column(String(50), unique=True)
    
    
class Booking(Base):
    __tablename__ = "bookings"
    id = Column(Integer, primary_key=True)
    user_id = Column(Integer, ForeignKey("users.id"))
    place_id = Column(Integer, ForeignKey("places.id"))
    start_date = Column(Date)
    end_date = Column(Date)
    price_per_night = Column(Float)
    num_nights = Column(Integer)
    
    
class Place(Base):
    __tablename__ = "places"
    id = Column(Integer, primary_key=True)
    host_id = Column(Integer, ForeignKey("hosts.id"), unique=True, nullable=False)
    address = Column(String(50))
    city_id = Column(Integer, ForeignKey("cities.id"))
    
    
class Review(Base):
    __tablename__ = "reviews"
    id = Column(Integer, primary_key=True)
    booking_id = Column(Integer, ForeignKey("bookings.id"))
    rating = Column(SmallInteger)
    review_body = Column(Text)
    
    
class Host(Base):
    __tablename__ = "hosts"
    id = Column(Integer, primary_key=True)
    user_id = Column(Integer, ForeignKey("users.id"))

    
class City(Base):
    __tablename__ = "cities"
    id = Column(Integer, primary_key=True)
    name = Column(String(25))
    country_id = Column(Integer, ForeignKey("countries.id"))
    
    
class Country(Base):
    __tablename__ = "countries"
    id = Column(Integer, primary_key=True)
    country_code = Column(String(5))
    name = Column(String(56)) # The United Kingdom of Great Britain and Northern Ireland
```

Utworzenie tabel

```python
Base.metadata.create_all(engine)
```

Wizualizacja

```python
# Utworzenie grafu
graph = create_schema_graph(metadata=Base.metadata)

# Zapis do pliku .png
img_name = "db_schema.png"
graph.write_png(img_name)

# Przedstawienie grafu w Pythonie
plt.figure(figsize=(25, 25))
plt.imshow(mpimg.imread(img_name))
plt.axis(False)
plt.show()
```

*Baza danych oraz tabele zostały pomyślnie stworzone, co widać na powyższym grafie.*
