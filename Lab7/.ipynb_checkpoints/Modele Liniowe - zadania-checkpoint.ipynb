##  Regresja liniowa

Regresja liniowa to tworzenie modeli przedstawiających wielkość, której wartość chcemy przewidzieć (zmienna zależna) jako kombinację liniową predyktorów (zmiennych zależnych).

Struktura modelu ma postać

$$\tilde{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 +\ldots+ \beta_n x_n$$

lub w formie macierzowej 

$$\tilde{\mathbf{y}} = \mathbf{[1\  X ]} \beta$$

gdzie wektor wartości zmiennych zależnych jest przewidywany na podstawie macierzy danych i wektora współczynników $\beta$.

Współczynnik $\beta_0$ w nomentkalturze anglojęzycznej jest nazywany *intercept* ponieważ jest to punkt przecięcia się z osią $y$ gdy wszystkie predyktory są równe 0.

Aby wyznaczyć parametry modelu liniowego potrzebujemy wektora danych uczących, tj. znanych nam wartości zmiennej zależnej  $\mathbf{y}$ oraz odpowiadająxcej mu macierzy predyktorów $\mathbf{X}$. Macierz predyktorów nosi nazwę *design matrix*, i bardzo użytyeczne do jej wyznaczenia (w oparciu o dane w forme dataframe) jest pakiet [patsy](https://patsy.readthedocs.io/en/latest/quickstart.html), a w szczególności funkcja ```dmatrices```.


Model regresyjny ma mocne umocowanie statystyczne, jednak w ramach zajęć będziemy traktować go machinistycznie jak jedną z metod uczenia maszynowego. Do tworzenia takich modeli służy klasa ```LinearRegression```

Do analizy regresji liniowej wykorzystywać będziemy pakiet ```scikit-learn``` oraz zawarte w nim klasy, głównie w modułach ```linear_model```, ```metrics``` oraz ```model_selection```. 

### Dane uczące i dane testowe

Zazwyczaj rozwiązując problem nie mamy dostępu do dodatkowych danych testowych uzupełniających nasze dane uczące. Dlatego też, popularną praktyką jest podzielenie posiadanego zbioru danych w sposób losowy na dwa zbiory: uczący i testowy. Do tego służy funkcja ```train_test_split``` z modułu ```model_selection```.

##### Zmienną niezależną są godziny poświęcone na naukę, gdyż to wynik zależy w pewnym stopniu od nich, a nie na odwrót.

````python
scores.plot(x="Hours", y="Scores", kind="scatter")
plt.title("Scores against hours spent on learning")
plt.grid(), plt.show()

scores_X = scores.iloc[:, :-1].values
scores_y = scores.iloc[:, 1].values
scores_X_train, scores_X_test, scores_y_train, scores_y_test = model_selection.train_test_split(scores_X, scores_y,
                                                                                                test_size=0.3)
scores_lr = linear_model.LinearRegression().fit(scores_X_train, scores_y_train)

print(f"Współczynniki kombinacji liniowej predyktorów: {scores_lr.coef_}")
print(f"Punkt przecięcia się wykresu regresji liniowej z osią y: {scores_lr.intercept_}")
````

##### Znaczenie wskaźników
- Wskaźniki coef_ to współczynniki kombinacji liniowej predyktorów. W przypadku gdy dany jest tylko 1 predyktor, dostajemy 1 współczynnik, który jest współczynnikiem kierunkowym (tangensem kąta nachylenia) prostej regresji liniowej.
- Wskaźnik intercept_ oznacza punkt przecięcia się wykresu regresji liniowej z osią $y$, gdy wszystkie predyktory są równe 0.

````python
scores_y_pred = scores_lr.predict(scores_X_test)

err = (abs(scores_y_pred - scores_y_test)) / scores_y_test * 100
scores_comp = pd.DataFrame({"Dane testowe": scores_y_test, "Predykcje": scores_y_pred, "Błąd względny [%]": err})
scores_comp
````

##### Błąd względny różni się dla kolejnych punktów testowych - ma duże wahania.

````python
plt.figure()
plt.scatter(scores_X_test, scores_y_test, label="Test set")
plt.plot(scores_X_test, scores_y_pred, c="r", label="Predicted set")
plt.xlabel("Hours"), plt.ylabel("Scores"), plt.title("Scores against hours spent on learning")
plt.legend(), plt.grid(), plt.show()

print(f"Średni błąd bezwzględny: {metrics.mean_absolute_error(scores_y_test, scores_y_pred)}")
print(f"Błąd średniokwadratowy: {metrics.mean_squared_error(scores_y_test, scores_y_pred)}")
print(f"Pierwiastek błędu średniokwadratowego: {metrics.mean_squared_error(scores_y_test, scores_y_pred, squared=False)}")
````

##### Uzyskana regresja liniowa jest satysfakcjonująca, co można zauważyć zarówno na wykresie, jak i patrząc na wartości poszczególnych błędów. Łatwo znaleźć prostą dopasowującą się do danych.

### Regresja logistyczna

Naturalnym rozwojem modeli linowych, są uogólnione modele liniowe (*generalized linear models*). 

Ich przewodnią ideą jest to, że zamiast reprezentować zmienną zależną jako kombinację liniową niezależnych, staramy się stworzyć model liniowy pewnej funkcji zmiennej zależnej.

$$ f(y) =  \beta_0 + \beta_1 x_1 + \beta_2 x_2 +\ldots+ \beta_n x_n$$

wymaga się, aby funkcja ta była odwracalna, dzięki czemu 

$$ y = f^{-1} ( \beta_0 + \beta_1 x_1 + \beta_2 x_2 +\ldots+ \beta_n x_n)$$

Takie przekształcenie stosujemy wtedy, gdy zbiór wartości zmiennej zależnej nie jest całym zbiorem liczb rzeczywistych tylko jest w jakiś sposób ograniczony. Na przykład do przedziału $[0,1]$, do wartości dodatnich lub wartości całkowitych. W tens posób można wybrać funkcję $f$, której dziedziną jest zbiór wartości $y$ zaś obrazem jest zbiór liczb rzeczywistych.

Jedną z takich funkcji jest tzw. funkcja logistyczna, która odpowiada zbiorowi $[0,1]$ przez co często jest wykorzystywana do analizy prawdopodobieństwa przynależności do danej klasy, lub też inaczej jako klasyfikator.

W takiej sytuacji zmienne zależne powinny przyjmować wyłącznie wartości $0$ i $1$ co odpowiada temu czy dany punkt należy do interesującej klasy. Model logistyczny poznwala na określenie estymaty prawdopodobieństwa, czy punkt należy czy nie do danej klasy. 

Do tworzenia modeli logistycznyc służy klasa: ```LogisticRegression```

W przypadku naszych odfiltrowanych danych będą to kolumny: cp, restecg, fbs, sex, exang. Jako, że trzy ostatnie już zapisane są w formie 0 i 1 tylko cp i restecg wymagają tej operacji.

````python
print(f"Wymiary: {heart.shape}")
heart.head()
````

````python
heart.describe()
````

````python
heart_cleaned = heart.replace("?", np.nan)
heart_cleaned.drop(columns=["slope", "ca", "thal"], inplace=True)
heart_cleaned.dropna(inplace=True)
heart_cleaned.reset_index(drop=True, inplace=True)

# Zmiana nazwy kolumny 'num       ' na 'num'
heart_cleaned.rename(columns={"num       ": "num"}, inplace=True)

# Przeniesienie kolumny 'num' na koniec
cols = [col for col in heart_cleaned.columns if col != "num"] + ["num"]
heart_cleaned = heart_cleaned[cols]
heart_cleaned.head()
````

````python
# Wyznaczenie dummies
heart_dummies = pd.get_dummies(heart_cleaned, columns=["cp", "restecg"], prefix=["cp", "restecg"])

# Przeniesienie kolumny 'num' na koniec
dummies_cols = [col for col in heart_dummies.columns if col != "num"] + ["num"]
heart_dummies = heart_dummies[dummies_cols]

# Wyznaczenie macierzy X oraz wektora y
heart_X = heart_dummies.iloc[:, :-1].values
heart_y = heart_dummies.iloc[:, -1].values

# Podział zbiorów na treningowy i testowy
heart_X_train, heart_X_test, heart_y_train, heart_y_test = model_selection.train_test_split(heart_X, heart_y, test_size=0.2)

# Dopasowanie - regresja logistyczna
heart_lr = linear_model.LogisticRegression(max_iter=10000).fit(heart_X_train, heart_y_train)

# Predykcje
heart_y_pred = heart_lr.predict(heart_X_test)

# Zgodność
print(f"Średnie dopasowanie: {heart_lr.score(heart_X_test, heart_y_test) * 100}%")
heart_comp = pd.DataFrame({"Dane testowe": heart_y_test, "Predykcje": heart_y_pred, "Zgodność": heart_y_pred == heart_y_test})
heart_comp
````

##### Zgodność jest dość wysoka.

````python
print(f"Średni błąd bezwzględny: {metrics.mean_absolute_error(heart_y_test, heart_y_pred)}")
print(f"Błąd średniokwadratowy: {metrics.mean_squared_error(heart_y_test, heart_y_pred)}")
print(f"Pierwiastek błędu średniokwadratowego: {metrics.mean_squared_error(heart_y_test, heart_y_pred, squared=False)}\n")
print("Raport klasyfikacji")
d = metrics.classification_report(heart_y_test, heart_y_pred, target_names=["Class 1", "Class 2"], zero_division=1,
                                  output_dict=True)

for k, v in d.items():
    print(k)
    print(v)
    print()
````

##### Błędy są stostunkowo dopuszczalne. Regresja logistyczna działa dość zadowalająco.

````python
# Podział na zmienne numeryczne i kategorialne
numerical_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]
categorical_cols = [col for col in dummies_cols if col not in ["age", "trestbps", "chol", "thalach", "oldpeak", "num"]]

# Wyznaczenie dummies
heart_dummies_scaled = pd.get_dummies(heart_cleaned, columns=["cp", "restecg"], prefix=["cp", "restecg"])

# Przeniesienie kolumny 'num' na koniec
dummies_scaled_cols = [col for col in heart_dummies_scaled.columns if col != "num"] + ["num"]
heart_dummies_scaled = heart_dummies_scaled[dummies_scaled_cols]

# Podział zbiorów na treningowy i testowy
heart_train_scaled, heart_test_scaled = model_selection.train_test_split(heart_dummies_scaled, test_size=0.2)

# Wyznaczenie macierzy X oraz wektora y
heart_X_test_scaled = heart_test_scaled.iloc[:, :-1].values
heart_y_test_scaled = heart_test_scaled.iloc[:, -1].values

# Skalowanie
scaler = preprocessing.StandardScaler()
scaler.fit(heart_train_scaled[numerical_cols])

heart_X_train_numerical_scaled = scaler.transform(heart_train_scaled[numerical_cols])
heart_X_train_categorical = heart_train_scaled[categorical_cols].to_numpy()
heart_X_train_scaled = np.hstack((heart_X_train_categorical, heart_X_train_numerical_scaled))
heart_y_train_scaled = heart_train_scaled["num"].values

# Dopasowanie - regresja logistyczna
heart_scaled_lr = linear_model.LogisticRegression().fit(heart_X_train_scaled, heart_y_train_scaled)

# Predykcje
heart_y_pred_scaled = heart_lr.predict(heart_X_test_scaled)

# Zgodność
print(f"Średnie dopasowanie: {heart_lr.score(heart_X_test_scaled, heart_y_test_scaled) * 100}%")
heart_comp_scaled = pd.DataFrame({"Dane testowe": heart_y_test_scaled, "Predykcje": heart_y_pred_scaled,
                                  "Zgodność": heart_y_pred_scaled == heart_y_test_scaled})
heart_comp_scaled
````

##### Zgodność wydaje się być porównywalna do tej sprzed skalowania.
